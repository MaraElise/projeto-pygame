```python
# Perguntar qual o nome do jogador
# aposta valor
#quer apostar o total 


# printar uma frase Good Look Jogador com um emogi sorrindo 
# abaixo um emogi malvado dizendo que não é tão facil assim 

# O objetivo sera chegar a linha de chegada e para isso ele precisa andar 100 casas

# ele encontrara desafios no meio do caminho, então algumas variaveis serão randomicas outras serão pre definidas

# ele podera ter 10 tentativas(sera um jogo de sorte)

# o jogador devera escolher opções 



#obs: colocar uma regra que o numero randomico não pode ser maior qeu o numero de casas que andou
# inicio do jogo 
# o jogador vai clicar em enter e aparecera um numero random
```

```python
import time

print(f'''\U0001F3C6 O objetivo do jogo é você chegar a linha de chegada!
Chegando a linha de chegada o sua aposta será devolvida em dobro!
''')
time.sleep(2)
Jogador = input('Qual o seu nome?     ')
print(f'''
\U0001f600 Olá {Jogador}, vamos iniciar!
''')
time.sleep(2)
Valor_aposta =int(input('\U0001F911 Quanto quer apostar em dinheiro?     '))
print(f'''
\U0001F917 Boa sorte!!''')
time.sleep(2)
print(f'''
\U0001F62C Mas.... não será tão fácil assim {Jogador}, 
tudo depederá da sua sorte, 
você encontrará desafios no caminho 
e terá apenas 15 chances!!
''')

import random
Casas = 0
Min=1
Max=6
i=0
Start='S'

while i < 15 and Casas < 20 and Start == 'S':
    
    VR = random.randint(Min, Max)
    
    time.sleep(2)
    Start = input(f'Olá {Jogador} vamos rodar os dados? S = Sim ou N = Não:  ').upper()
    if Start != 'S' and Start != 'N':
        print('Essa opção não é aceita')
    else:
        if Start == 'S':
            time.sleep(2)
            print('Rodando os dados')
            i = i + 1
            time.sleep(1)
            print(f'Você ainda tem {15-i} tentativas')
            if Casas >=20:
                time.sleep(2)
                print(f'''*****{Jogador} você ganhou! 
    Andou as 20 casas, no limite de tentativas! 
    Ganhou R$ {Valor_aposta*2}!!!! \U0001F929''')
            else:
                if Casas == 0:
                    Casas = Casas + VR
                    time.sleep(2)
                    print(f'Você andou um total de {Casas} casas')
                else:              
                    if VR % 2 == 0:
                        Casas = Casas + VR
                        time.sleep(2)
                        print(f'Uauu você vai andar + {VR} casas')
                        print(f'Isso ai já andou um total de {Casas } casas')
                    else:
                        time.sleep(2)
                        print(f'Que Pena tera que voltar algumas casas!!\nEspero que não sejam todas as casas!!')
                        if  VR >= Casas:
                            Casas = ((Casas) *(-1))
                            time.sleep(2)
                            print(f'Não foi dessa vez você voltou todas as casas')
                        else: 
                            Casas = Casas - VR
                            time.sleep(2)
                            print(f'Você voltara - {VR} casas')
                            print(f'Não fique triste, mesmo assim você já andou {Casas} casas')
                    
                    
               
            
            
            
        else:
            time.sleep(2)
            print(f'{Jogador}, o jogo foi encerrado, você escolheu não continuar.\U0001F97A')
if Casas >=20:
    time.sleep(2)
    print(f'''*****{Jogador} você ganhou! 
    Andou as 20 casas, no limite de tentativas! 
    Ganhou R$ {Valor_aposta*2}!!!! \U0001F929''')
elif Start != 'S':
    time.sleep(2)
    print('Até mais Tarde!\U0001F609')
else:
    time.sleep(2)
    print(f'''*****{Jogador} você perdeu todo seu dinheiro!!\U0001F616 
    Usou todas as suas tentativas!  \U0001F633''')
```
