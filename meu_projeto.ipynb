```python
import time

print(f'''\U0001F3C6 O objetivo do jogo Ã© vocÃª chegar a linha de chegada!
Chegando a linha de chegada o sua aposta serÃ¡ devolvida em dobro!
''')
time.sleep(2)
Jogador = input('Qual o seu nome?     ')
print(f'''
\U0001f600 OlÃ¡ {Jogador}, vamos iniciar!
''')
time.sleep(2)
Valor_aposta =int(input('\U0001F911 Quanto quer apostar em dinheiro?     '))
print(f'''
\U0001F917 Boa sorte!!''')
time.sleep(2)
print(f'''
\U0001F62C Mas.... nÃ£o serÃ¡ tÃ£o fÃ¡cil assim {Jogador}, 
tudo depederÃ¡ da sua sorte, 
vocÃª encontrarÃ¡ desafios no caminho 
e terÃ¡ apenas 15 chances!!
''')

import random
Casas = 0
Min=1
Max=6
i=0
Start='S'

while i < 15 and Casas < 20 and Start == 'S':
    
    VR = random.randint(Min, Max)
    
    time.sleep(2)
    Start = input(f'OlÃ¡ {Jogador} vamos rodar os dados? S = Sim ou N = NÃ£o:  ').upper()
    if Start != 'S' and Start != 'N':
        print('Essa opÃ§Ã£o nÃ£o Ã© aceita')
    else:
        if Start == 'S':
            time.sleep(2)
            print('Rodando os dados')
            i = i + 1
            time.sleep(1)
            print(f'VocÃª ainda tem {15-i} tentativas')
            if Casas >=20:
                time.sleep(2)
                print(f'''*****{Jogador} vocÃª ganhou! 
    Andou as 20 casas, no limite de tentativas! 
    Ganhou R$ {Valor_aposta*2}!!!! \U0001F929''')
            else:
                if Casas == 0:
                    Casas = Casas + VR
                    time.sleep(2)
                    print(f'VocÃª andou um total de {Casas} casas')
                else:              
                    if VR % 2 == 0:
                        Casas = Casas + VR
                        time.sleep(2)
                        print(f'Uauu vocÃª vai andar + {VR} casas')
                        print(f'Isso ai jÃ¡ andou um total de {Casas } casas')
                    else:
                        time.sleep(2)
                        print(f'Que Pena tera que voltar algumas casas!!\nEspero que nÃ£o sejam todas as casas!!')
                        if  VR >= Casas:
                            Casas = ((Casas) *(-1))
                            time.sleep(2)
                            print(f'NÃ£o foi dessa vez vocÃª voltou todas as casas')
                        else: 
                            Casas = Casas - VR
                            time.sleep(2)
                            print(f'VocÃª voltara - {VR} casas')
                            print(f'NÃ£o fique triste, mesmo assim vocÃª jÃ¡ andou {Casas} casas')
                    
                    
               
            
            
            
        else:
            time.sleep(2)
            print(f'{Jogador}, o jogo foi encerrado, vocÃª escolheu nÃ£o continuar.\U0001F97A')
if Casas >=20:
    time.sleep(2)
    print(f'''*****{Jogador} vocÃª ganhou! 
    Andou as 20 casas, no limite de tentativas! 
    Ganhou R$ {Valor_aposta*2}!!!! \U0001F929''')
elif Start != 'S':
    time.sleep(2)
    print('AtÃ© mais Tarde!\U0001F609')
else:
    time.sleep(2)
    print(f'''*****{Jogador} vocÃª perdeu todo seu dinheiro!!\U0001F616 
    Usou todas as suas tentativas!  \U0001F633''')
```
