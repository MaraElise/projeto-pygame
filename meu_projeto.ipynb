```python
import time

print(f'''\U0001F3C6 O objetivo do jogo √© voc√™ chegar a linha de chegada!
Chegando a linha de chegada o sua aposta ser√° devolvida em dobro!
''')
time.sleep(2)
Jogador = input('Qual o seu nome?     ')
print(f'''
\U0001f600 Ol√° {Jogador}, vamos iniciar!
''')
time.sleep(2)
Valor_aposta =int(input('\U0001F911 Quanto quer apostar em dinheiro?     '))
print(f'''
\U0001F917 Boa sorte!!''')
time.sleep(2)
print(f'''
\U0001F62C Mas.... n√£o ser√° t√£o f√°cil assim {Jogador}, 
tudo depeder√° da sua sorte, 
voc√™ encontrar√° desafios no caminho 
e ter√° apenas 15 chances!!
''')

import random
Casas = 0
Min=1
Max=6
i=0
Start='S'

while i < 15 and Casas < 20 and Start == 'S':
    
    VR = random.randint(Min, Max)
    
    time.sleep(2)
    Start = input(f'Ol√° {Jogador} vamos rodar os dados? S = Sim ou N = N√£o:  ').upper()
    if Start != 'S' and Start != 'N':
        print('Essa op√ß√£o n√£o √© aceita')
    else:
        if Start == 'S':
            time.sleep(2)
            print('Rodando os dados')
            i = i + 1
            time.sleep(1)
            print(f'Voc√™ ainda tem {15-i} tentativas')
            if Casas >=20:
                time.sleep(2)
                print(f'''*****{Jogador} voc√™ ganhou! 
    Andou as 20 casas, no limite de tentativas! 
    Ganhou R$ {Valor_aposta*2}!!!! \U0001F929''')
            else:
                if Casas == 0:
                    Casas = Casas + VR
                    time.sleep(2)
                    print(f'Voc√™ andou um total de {Casas} casas')
                else:              
                    if VR % 2 == 0:
                        Casas = Casas + VR
                        time.sleep(2)
                        print(f'Uauu voc√™ vai andar + {VR} casas')
                        print(f'Isso ai j√° andou um total de {Casas } casas')
                    else:
                        time.sleep(2)
                        print(f'Que Pena tera que voltar algumas casas!!\nEspero que n√£o sejam todas as casas!!')
                        if  VR >= Casas:
                            Casas = ((Casas) *(-1))
                            time.sleep(2)
                            print(f'N√£o foi dessa vez voc√™ voltou todas as casas')
                        else: 
                            Casas = Casas - VR
                            time.sleep(2)
                            print(f'Voc√™ voltara - {VR} casas')
                            print(f'N√£o fique triste, mesmo assim voc√™ j√° andou {Casas} casas')
                    
                    
               
            
            
            
        else:
            time.sleep(2)
            print(f'{Jogador}, o jogo foi encerrado, voc√™ escolheu n√£o continuar.\U0001F97A')
if Casas >=20:
    time.sleep(2)
    print(f'''*****{Jogador} voc√™ ganhou! 
    Andou as 20 casas, no limite de tentativas! 
    Ganhou R$ {Valor_aposta*2}!!!! \U0001F929''')
elif Start != 'S':
    time.sleep(2)
    print('At√© mais Tarde!\U0001F609')
else:
    time.sleep(2)
    print(f'''*****{Jogador} voc√™ perdeu todo seu dinheiro!!\U0001F616 
    Usou todas as suas tentativas!  \U0001F633''')
```
